# Classifying Iris flowers using KNN algorithm

**Importing Libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
import seaborn as sns
from sklearn.model_selection import train_test_split
import numpy as np
```

**Loading Dataset**

```python
dataset = pd.read_csv('Iris.csv')
```

**Extracting dependant and independant variables**

```python
x = dataset[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
y = dataset[['Species']]
```

**Visualizing the data**

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Species', data=y, hue='Species', palette='Set1', legend=False)
plt.title('Count of Species')
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()
```

**Spliting data into training and test set**

```python
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size = 1/3, random_state=0)
```

**Training the model**

```python
KNN = KNeighborsClassifier(n_neighbors = 7, metric = 'minkowski', p = 2)
train_y = train_y['Species']
KNN.fit(train_x, train_y)
```

**Predicting the outcomes of testing set**

```python
predicted_y = KNN.predict(test_x)
predicted_y
```

**Measuring the accuracy of model**

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(test_y, predicted_y)
```
